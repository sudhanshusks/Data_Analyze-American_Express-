```python
test= pd.read_csv("data/Leaderboard_dataset.csv")
test= test.replace('missing', np.nan)
test= test.replace('na', np.nan)
test.isnull().sum()
```

```python
test[na_cols]= test[na_cols].astype(str).astype(float)
test[missing_cols]= test[missing_cols].astype(str).astype(float)

test['mvar47']= le.fit_transform(test['mvar47'])

test= test.fillna(test.median())
test.head()
```

```python
Xtest= test.drop(['application_key'], axis=1)
```

```python
ids= test['application_key']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
clf = RandomForestClassifier()
scores = cross_val_score(clf, X, y)
print(scores.mean())

classifier = clf.fit(X,y)
predictions = classifier.predict_proba(Xtest)
print(predictions)
```

```python
data= dict({'application_key' : ids, 'predictions-0':predictions[:,0]})
f= pd.DataFrame(data)
f.to_csv("data/temp.csv")
```

```python
lr = LogisticRegression()
scores = cross_val_score(lr, X, y)
print(scores.mean())
classifier = lr.fit(X,y)
predictions = classifier.predict_proba(Xtest)

data= dict({'application_key' : ids, 'predictions-0':predictions[:,0]})
f= pd.DataFrame(data)
```

```python
f= f.sort_values('predictions-0')
f
```

```python
nb = GaussianNB()
scores = cross_val_score(nb, X, y)
print(scores.mean())
classifier = nb.fit(X,y)
predictions = classifier.predict_proba(Xtest)
```

```python
svm = svm.SVC()
scores = cross_val_score(svm, X, y)
print(scores.mean())
classifier = svm.fit(X,y)
predictions = classifier.predict_proba(Xtest)
```

```python
pred=[]
for i in range(len(predictions)):
    if(predictions[i,1] >0.85):
        pred.append(1)
    else:
        pred.append(0)
```
